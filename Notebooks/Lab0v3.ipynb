```python
# Ejemplo de break y continue
for value in [3, 1, 4, 1, 5, 9, 2]:
    print("Checking", value)
    if value > 8:
        print("Exiting for loop")
        break
    elif value < 3:
        print("Ignoring")
        continue
    print("The square is", value**2)
```
